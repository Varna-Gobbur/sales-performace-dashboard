```python
# Fix date columns
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Add useful columns
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.to_period("M").astype(str)
df["Profit Margin"] = (df["Profit"] / df["Sales"]).round(4)
df["Days to Ship"] = (df["Ship Date"] - df["Order Date"]).dt.days

# Remove duplicates
df = df.drop_duplicates(subset="Order ID")

# Check for nulls
print("Null values:\n", df.isnull().sum())
print(f"\nClean dataset: {len(df)} rows")
```

```python
total_revenue   = df["Sales"].sum()
total_profit    = df["Profit"].sum()
avg_margin      = df["Profit Margin"].mean() * 100
total_orders    = df["Order ID"].nunique()
avg_order_value = total_revenue / total_orders

print(f"Revenue:          ${total_revenue:,.0f}")
print(f"Profit:           ${total_profit:,.0f}")
print(f"Avg Margin:       {avg_margin:.1f}%")
print(f"Total Orders:     {total_orders:,}")
print(f"Avg Order Value:  ${avg_order_value:,.0f}")
```

```python
base = "/Users/varna/Desktop/Personal/Portfolio/superstore_dataset/"

df.to_csv(base + "sales_clean.csv", index=False)

monthly = df.groupby("Month")["Sales"].sum().reset_index()
monthly.to_csv(base + "monthly_sales.csv", index=False)

region = df.groupby("Region")[["Sales","Profit"]].sum().reset_index()
region.to_csv(base + "region_summary.csv", index=False)

category = df.groupby("Category")[["Sales","Profit"]].sum().reset_index()
category.to_csv(base + "category_summary.csv", index=False)

print("All files exported!")
```
